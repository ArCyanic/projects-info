```python
import pandas as pd
```

```python
dst = 'TOP20.csv'
something = ['patentsNumber', 'inventor', 'referNumber', 'referedNumber']
# type : '专利权人'等
# areaType : '城市', '省份'
head = ['id', 'areaType', 'type', 'area', 'year', 'value', 'DeleteStatus']


def mergeCSV(dst, head):
    result = pd.DataFrame(columns=head)

    areaTypes = ['城市', '省份']
    properties = ['专利数量', '发明人', '引用', '被引用']
    
    id = 0

    for areaType in areaTypes:
        for property in properties:
            path = 'required/量子TOP20/{}量子{}TOP20.csv'.format(areaType, property)
            df = pd.read_csv(path, index_col='{}'.format(areaType))
            # 对于每一个文件
            for area in df.index:
                for year in [str(i) for i in range(2000, 2020 + 1)]:
                    id += 1
                    # 对于每一个元素，都在数据库中新建一行
                    newRow = {
                        'id' : id, 
                        'areaType' : areaType, 
                        'type' : property, 
                        'area' : area, 
                        'year' : year, 
                        'value' : df.loc[area, year], 
                        'DeleteStatus' : 0
                    }
                    result = result.append(newRow, ignore_index=True)
    result.to_csv(dst, index=False)
    return result

mergeCSV(dst, head)
                    
                
```

```python
def transformCSV_1(src, dst , head):
    df = pd.read_csv(src)
    df.index = df.iloc[:, 0]
    df = df.drop(columns=df.columns[0])

    # 首先提取出列名
    colsName = df.columns
    # 提取出行首
    rowsName = df.index
    result = pd.DataFrame(columns=head)
    id = 1

    for colName in colsName:
        for rowName in rowsName:
            temp = df.loc[rowName, colName]
            d = {key : val for key, val in zip(head, [id, colName, rowName, temp, 0])}
            # print(d)
            result = result.append(d, ignore_index=True)
            id += 1
    print(result)
    result.to_csv(dst, index=False)

src = 'resource/技术功效趋势.csv'
head = ['id', 'year', 'TechnicalEfficacy', 'value', 'DeleteStatus']
transformCSV_1(src, '技术功效趋势.csv', ['id', 'year', 'TechnicalEfficacy', 'value', 'DeleteStatus'])
```
